**Merging all cleaned data**

```python
gold = pd.read_csv(r"C:\Users\lenovo\Desktop\shivu\gold-currency-dynamics\data\cleaned\gold_clean.csv", parse_dates=["date"])
forex = pd.read_csv(r"C:\Users\lenovo\Desktop\shivu\gold-currency-dynamics\data\cleaned\forex_cleaned.csv", parse_dates=["date"])
fed = pd.read_csv(r"C:\Users\lenovo\Desktop\shivu\gold-currency-dynamics\data\cleaned\cleaned_fed_intrest.csv", parse_dates=["date"])
inflation = pd.read_csv(r"C:\Users\lenovo\Desktop\shivu\gold-currency-dynamics\data\cleaned\cleaned_cpi_inflation.csv", parse_dates=["date"])
```

```python
#Checking
print(f"Gold:----\n{gold.head()}")
print(f"Forex:----\n{forex.head()}")
print(f"Fed:----\n{fed.head()}")
print(f"CPI/Inflation:----\n{inflation.head()}")
```

```python
#Converting YYYY-MM-DD to YYYY-MM for gold and forex as gold dates are not standard
gold["month_key"] = gold["date"].dt.to_period("M")
forex["month_key"] = forex["date"].dt.to_period("M")
```

```python
#Recheck
print(f"Gold:----\n{gold.head()}")
print(f"Forex:----\n{forex.head()}")
```

```python
#Merging gold and forex
master = pd.merge(gold, forex, on='month_key', how='inner')
```

```python
#Dropping non used columns, Rename and aliging columns
master = master.drop(["date_x", "year", "month"], axis=1)
master = master.rename(columns={"date_y":"date"})
master = master[["date", "price_usd", "inr_per_usd", "month_key"]]
```

```python
master["gold_price_inr"] = master["price_usd"] * master["inr_per_usd"]
```

```python
master.head()
```

```python
fed['month_key'] = fed['date'].dt.to_period("M")
inflation['month_key'] = inflation['date'].dt.to_period("M")
```

```python
print(fed.head())
print(inflation.head())
```

```python
fed_monthly = fed.drop_duplicates(subset="month_key", keep="last")[['month_key', 'fed_intrest_rate']]
inflation_monthly = inflation.drop_duplicates(subset="month_key", keep="last")[["month_key", "inflation_yoy"]]
```

```python
print(fed_monthly.head())
print(inflation_monthly.head())
```

```python
#Merge fed and inflation with master
master = master.merge(fed_monthly, on='month_key', how='left')
master = master.merge(inflation_monthly, on="month_key", how="left")
```

```python
#dropping month_key column
master = master.drop(columns=["month_key"])
```

```python
#Adding calculated columns
master["gold_return_1d"] = master["price_usd"].pct_change() * 100
master["inr_change_1d"] = master["inr_per_usd"].pct_change() * 100
master["gold_inr_return_1d"] = master["gold_price_inr"].pct_change() * 100
```
